## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 1. Remove duplicates
data = data.drop_duplicates()

# 2. Standardize column names
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')

# 3. Handle unrealistic ages
data = data[(data['age'] >= 15) & (data['age'] <= 80)].copy()  # create safe copy

# 4. Clean and normalize gender values
data.loc[:, 'gender'] = data['gender'].str.lower().str.strip()
data.loc[:, 'gender'] = data['gender'].replace({
    'male': 'male', 'm': 'male', 'man': 'male', 'cis male': 'male',
    'female': 'female', 'f': 'female', 'woman': 'female', 'cis female': 'female'
})
data.loc[:, 'gender'] = data['gender'].apply(lambda x: x if x in ['male', 'female'] else 'other')

# 5. Fill missing values in selected categorical fields
cat_cols = ['self_employed', 'work_interfere', 'state', 'benefits',
            'care_options', 'wellness_program', 'seek_help', 'anonymity',
            'leave', 'mental_health_consequence', 'phys_health_consequence',
            'coworkers', 'supervisor', 'mental_health_interview',
            'phys_health_interview', 'mental_vs_physical', 'obs_consequence']

for col in cat_cols:
    data.loc[:, col] = data[col].fillna('Unknown')

# 6. Strip whitespace from all string columns
str_cols = data.select_dtypes(include='object').columns
data.loc[:, str_cols] = data[str_cols].apply(lambda x: x.str.strip())

# 7. Reset index
data.reset_index(drop=True, inplace=True)

# Final check
print("Data is cleaned now and ready for analysis.")
print(data.info())
````

### What all manipulations have you done and insights you found?

What I did during data wrangling:
Removed duplicate rows to ensure data consistency.

Standardized column names by converting them to lowercase and replacing spaces with underscores.

Filtered out unrealistic age values (e.g., age < 15 or age > 80) to maintain valid demographic information.

Cleaned and normalized the gender column, grouping various text responses into “male,” “female,” or “other.”

Filled missing values in categorical columns such as self_employed, benefits, care_options, etc., using 'Unknown' to ensure consistency.

Stripped extra whitespace from all string columns for better matching and consistency.

Reset the index to reflect changes made during filtering and cleaning.

Insights found after cleaning:
Most respondents are aged between 21 and 35, indicating mental health feedback is primarily from early-to-mid-career professionals.

A large number of gender entries were inconsistent, requiring consolidation to improve demographic analysis.

Many fields had missing values, especially in workplace-related questions like benefits, wellness_program, and care_options, which may reflect a lack of communication or support from employers.

Stigma-related fields like mental_health_consequence and obs_consequence had high levels of concern, showing that fear of negative outcomes still affects openness about mental health.

Remote workers and employees in larger companies appeared to report better access to resources and more openness in discussing mental health.

#### Chart - 14 - Correlation Heatmap

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load your dataset (replace with actual file path if using a file)
data = pd.read_csv("/content/survey.csv")  # Replace with your CSV file if needed

# Encode categorical columns
df_encoded = data.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col].astype(str))

# Compute the correlation matrix
correlation_matrix = df_encoded.corr()

# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Heatmap")
plt.show()

````
